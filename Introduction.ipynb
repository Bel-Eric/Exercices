**Pour aller plus loin :**
*   Comptez le nombre de tentatives et affichez-le dans le message final.
*   Donnez un nombre d'essais limité (par exemple 10) et faites perdre le joueur s'il ne trouve pas le nombre à temps.

````python
from random import randint

aleatoire= randint(1,100)

#j'ai utiliser la boucle for la place de while et ai arrete la boucle a 3 tentative max
for tentative in range(3):
  print(f"Essais {tentative+1}/3 \n")
  trouver=False
  try:

    choix_joueur=int(input("deviner le nombre du systeme entre 1 et 100:  "))

    if choix_joueur < 1 or choix_joueur > 100:

      print(" \t vous dever choisir un nombre entre 1 et 100")
      continue

    if choix_joueur < aleatoire :

      print(f"ce nombre est trop pétit, {("\nVeiller reessayer:") if tentative<2 else ""} ")

    elif choix_joueur> aleatoire:

      print(f"ce nombre est trop grand.{("\nVeiller reessayer:") if tentative<2 else ""} ")

    else:

      trouver=True
      print(f"Bravo, vous l'avez trouvé a {tentative+1} tantative")
      break


  except ValueError :
    print("veillez saisir un nombre entier")

if not trouver:
  print("vous avez epuiser vos temps d'essai")
````

### Exercice 7 : Mini-Projet - Inventaire de Magasin (Très Avancé)

Ce projet final combine tout ce que vous avez appris : fonctions, listes, dictionnaires, boucles, conditions et gestion des erreurs. L'objectif est de créer un système de gestion d'inventaire pour un petit magasin.

**Structure des données :**
Votre inventaire sera une **liste de dictionnaires**. Chaque dictionnaire représentera un produit et aura les clés suivantes : `"nom"` (str), `"prix"` (float), `"quantite"` (int).

```python
# Exemple de structure de données
inventaire = [
    {"nom": "Pomme", "prix": 0.5, "quantite": 100},
    {"nom": "Banane", "prix": 0.3, "quantite": 150},
    {"nom": "Orange", "prix": 0.6, "quantite": 80}
]
```

**Instructions :**
Créez un programme complet structuré autour d'un menu principal dans une boucle `while True`. Le programme doit proposer les options suivantes, chacune implémentée dans sa propre fonction.

1.  **`afficher_inventaire(inv: list) -> None:`**
    *   Affiche tous les produits de l'inventaire sous forme de tableau bien formaté.
    *   Calcule et affiche la valeur totale de l'inventaire (somme de `prix * quantite` pour chaque produit).

2.  **`ajouter_produit(inv: list) -> None:`**
    *   Demande à l'utilisateur le nom, le prix et la quantité du nouveau produit.
    *   **Validation :** Utilisez `try-except` pour vous assurer que le prix et la quantité sont bien des nombres.
    *   **Logique :** Vérifiez si un produit du même nom existe déjà. Si oui, proposez de mettre à jour la quantité. Sinon, ajoutez le nouveau produit (dictionnaire) à la liste d'inventaire.

3.  **`rechercher_produit(inv: list) -> None:`**
    *   Demande un nom de produit à l'utilisateur.
    *   Parcourt l'inventaire pour trouver le produit.
    *   Si trouvé, affiche ses détails. Sinon, affiche un message d'erreur.

4.  **`supprimer_produit(inv: list) -> None:`**
    *   Demande un nom de produit à l'utilisateur.
    *   Si le produit existe, le supprime de la liste. Affichez une confirmation. (Indice : `list.remove()` fonctionne sur les dictionnaires, mais il faut d'abord trouver le bon dictionnaire à supprimer).
    *   Sinon, affiche un message d'erreur.

5.  **Le programme principal :**
    *   Initialisez la liste `inventaire` avec quelques produits.
    *   Créez une boucle `while True` qui affiche un menu (1. Afficher, 2. Ajouter, 3. Rechercher, 4. Supprimer, 5. Quitter).
    *   En fonction du choix de l'utilisateur, appelez la fonction correspondante.
    *   Si l'utilisateur choisit de quitter, sortez de la boucle avec `break`.

````python
inventaire = [
    {"nom": "Pomme", "prix": 0.5, "quantite": 100},
    {"nom": "Banane", "prix": 0.3, "quantite": 150},
    {"nom": "Orange", "prix": 0.6, "quantite": 80}
]

def afficher_inventaire(inv : list)-> None:
  somme=0
  if not inv:
    print("Vous n'avez pas de produit enregistrer. ")

  else:

    for produit in inv:

      print(f"""
                Nom\t Prix \t Quantite
                {produit['nom']}\t {produit['prix']}£\t {produit['quantite']}""")
      somme= somme + (produit['prix'] * produit['quantite'])
  print(f"la somme total de l'inventaire est: { somme}£")

def ajouter_produit(inv : list) -> None :
  nom_produit= input("Saisir le  nom du produit")

  for produit in inv:

    if produit['nom'] == nom_produit:
      while True:
        try:
            M_A_quantite=int(input("Ce produit existe deja.\n Saisir une quantite pour faire une mise à jour du0e stock, si non saisir 0  "))
            break
        except ValueError:
          print("vous dever saisir un nombre")

      match M_A_quantite:

        case 0:
          print("Retour au menu principal.")
          break
        case _:
          produit['quantite']=M_A_quantite
      break


  else:
    while True:

      try:

        prix=float(input("saisir le prix du nouveau produit"))
        quantite= int(input("saisir la quantite du nouveau produit"))
        break

      except ValueError:

        print("les valeur saisies doivent etre des nombre.")
    inv.append({'nom' : nom_produit, 'prix' : prix, 'quantite' : quantite})
    print("Ajout ...")

def recherche_produit(inv : list) -> None:

  produit_recherche = input("Saisir le nom du produit rechercher: ")

  for produit in inv:

    if produit["nom"] == produit_recherche:
      print(f"""
                Nom \t Prix \t Quantite
                {produit['nom']} \t {produit["prix"]} \t {produit["quantite"]} """)
      break
  else:
    print("Ce produit n'est pas dans votre inventaire.")
    while True:

      try:

        quantite= int(input("saisir la quantite du nouveau produit pour l'ajouter, si non saisir 0 "))
        if quantite == 0:

          print("Retour au menu principal ...")
          break

        else:

          prix = float(input("saisir le prix du produit "))
          inv.append({'nom' : produit_recherche , 'prix' : prix , 'quantite' : quantite })
        break

      except ValueError:

        print("vous devez saisir un nombre")

def supprimer_produit(inv: list):

  produit_a_supprimer = input(" saisir le nom du produit a supprimer ")
  for num, produit in enumerate(inv):
    if produit['nom'] == produit_a_supprimer:
      inv.pop(num)
      print("Produit supprimer.")
      break
  else:
   print("ce produit n'est pas dans l'inventaire..")

def affiche_information() -> None :

  while True:
    try:
      choix = int(input("""
              Veiller saisir un nombre pour une operation:
              1 - Afficher Inventaire
              2 - Ajouter Produit
              3 - Rechercher Produit
              4 - Supprimer un produit
              0 - Rien faire"""))
    except ValueError:
      print("veiller saisir un nombre")
      continue
    match choix:
      case 1:
        afficher_inventaire(inventaire)

      case 2:
        ajouter_produit(inventaire)
      case 3:
        recherche_produit(inventaire)
      case 4:
        supprimer_produit(inventaire)
      case 0:
        break
      case _ :
        print("veiller saisir un nombre compris entre 0 et 4")

## le corp du programme
affiche_information()
````
