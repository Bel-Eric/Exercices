#### **Exercice 1 : Créer une classe `Livre` (Facile)**

**Objectif :** Maîtriser la syntaxe de base d'une classe, le constructeur `__init__`, les attributs et une méthode simple.

**Contexte :**
Vous devez modéliser un livre. Un livre est défini par un titre, un auteur et un nombre de pages. On doit pouvoir afficher une description simple du livre.

**Instructions :**
1.  Créez une classe nommée `Livre`.
2.  Dans le constructeur `__init__`, initialisez les attributs : `titre`, `auteur`, et `nombre_pages`.
3.  Créez une méthode `description()` qui ne prend pas de paramètres (à part `self`) et qui affiche une chaîne de caractères formatée comme suit : `"<Titre>" par <Auteur> contient <nombre_pages> pages.`
4.  Instanciez deux objets de la classe `Livre` avec les données de votre choix.
5.  Appelez la méthode `description()` sur chacun de vos objets.

```python
class Livre:
  """
    permet la creation et l'affichage des information d'un livre.

    un produit est caratérisé par:
      - nom_auteur : str
          le nom de l'auteur du livre,
      - titre_livre : str
          le titre du livre,
      - nombre_page : int
          le nombre de page du livre,
      - annee_publication : int
          l'anneé de publication du livre
    Important:
      le nombre de livre et l'année de publication doivent etre des entier supperieur a 0.
      si non ils sont representer par un chaine de caractere (" inconnu ")

  """

  def __init__( self,nom_auteur : str , titre_livre : str , nombre_page : int , annee_publication : int ) -> None :

    self.nom_auteur= nom_auteur
    self.titre_livre = titre_livre
    self.nombre_page = nombre_page if isinstance( nombre_page , int) and nombre_page >= 0 else "inconnu"
    self.annee_publication = annee_publication if isinstance( annee_publication , int) and annee_publication >= 0  else "inconnu"

  def Description_livre(self) -> None :
    """
      Donne la dèscription d'un produit.

      Paramètre:
      ---------
      Pas de paramètre

      Returns:
      -------
      None
    """

    print(f"""
               Livre:                   {self.titre_livre}
               Auteur:                  {self.nom_auteur}
               Annee de publication:    { self.annee_publication}
               Nombre de page:          {self.nombre_page} \n""")

Livre_1 = Livre( "Albert CAMUS" , "L'etranger" , 150 , 1974)
Livre_2 = Livre( "Koffi Inadjo" , "Sur les route sanglante...",  "mon nom", -50)
Livre_3 = Livre( "Alpha Blondy" ,"Jerusalem" , 150,  2007)

Livre_1.Description_livre()
Livre_2.Description_livre()
Livre_3.Description_livre()
```
