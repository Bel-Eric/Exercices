### Exercice 1 : Le calculateur de pourboire (Facile)

L'objectif est de créer un petit programme qui calcule le montant total à payer dans un restaurant, incluant le pourboire.

**Instructions :**
1.  Demandez à l'utilisateur de saisir le montant de l'addition (un nombre à virgule). Utilisez la fonction `input()`.
2.  Demandez-lui ensuite de saisir le pourcentage de pourboire qu'il souhaite laisser (un nombre entier, par exemple 15 pour 15%).
3.  **Attention :** N'oubliez pas de convertir les valeurs reçues de `input()` (qui sont des `str`) en types numériques appropriés (`float` et `int`).
4.  Calculez le montant du pourboire.
5.  Calculez le montant total de l'addition (addition + pourboire).
6.  Affichez les résultats de manière claire et formatée en utilisant une **f-string**. Par exemple :
    ```
    Pour une addition de 50.0€ et un pourboire de 15%:
    - Montant du pourboire : 7.5€
    - Montant total à payer : 57.5€
    ```

````python
while(True):
  try:

    Consomation= float( input(" Quel est le montant de votre consomation ? ") )
    Pourcentage= int( input(" Combien de pourcent de l'addition lassez en pourboir ? ") )
    break

  except ValueError:
      print("vous dever saisir des nombres")
Total=Consomation * (1+ ( Pourcentage/100 ) )
print(f"pour une consomation egale a {Consomation}£, et un pourboir de {Pourcentage}%,")
print(f"le total de votre payement est: {Total}£")

````

Votre inventaire sera une **liste de dictionnaires**. Chaque dictionnaire représentera un produit et aura les clés suivantes : `"nom"` (str), `"prix"` (float), `"quantite"` (int).

```python

]
```

5.  **Le programme principal :**
